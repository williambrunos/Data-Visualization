# 2.

Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10, 10))
sns.barplot(y='Film', x='Budget_M', data=df, edgecolor='0.2')
plt.title('Orçamento de todos os filmes (Milhões de Dólares)')
plt.ylabel('Filmes')
plt.xlabel('Orçamento do Filme')
plt.show()
```

```python
df[df['Budget_M'] == df['Budget_M'].max()]
```

```python
df[df['Budget_M'] == df['Budget_M'].min()]
```

Portanto, pelo gráfico gerado é possível observar que o filme com **maior orçamento** foi "Piratas do Caribe: no Fim do Mundo" e com **menor orçamento** foi "Crepúsculo: Lua Nova".

# 3.

Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

```python
plt.figure(figsize=(10, 10))
sns.barplot(y='Film', x='Worldwide_Gross_M', data=df, edgecolor='0.2')
plt.title('Bilheteria de todos os filmes (Milhões de Dólares)')
plt.ylabel('Filmes')
plt.xlabel('Bilheteria')
```

```python
df[df['Worldwide_Gross_M'] == df['Worldwide_Gross_M'].max()]
```

```python
df[df['Worldwide_Gross_M'] == df["Worldwide_Gross_M"].min()]
```

Portanto, a partir da visualização acima, percebemos que o filme com **maior bilheteria** foi Avatar, e a com **menor bilheteria** foi Ratatouille (de acordo com a célula de código anterior)

# 4.

Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano .
Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

```python
plt.figure(figsize=(15, 10))
sns.countplot(x='Year', hue='Genre', data=df, palette='pastel', edgecolor='0.2')
plt.title('Contagem de filmes por gênero em cada ano')
plt.ylabel('Contagem')
plt.xlabel('Anos')
plt.show()
```

Portanto, pela visualização acima é fácil perceber que os anos com **menos filmes de ação** foram 2008 e 2010 e com **menos filmes de animação** foram os anos 2008 e 2011.

# 5.

Crie uma visualização que mostre a média de orçamento de filmes por gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento?

```python
plt.figure(figsize=(15, 10))
sns.barplot(x='Year', y='Budget_M', hue='Genre', palette='pastel', edgecolor='0.2', data=df, ci=None)
plt.title('Média de orçamento de filmes por gênero em cada ano')
plt.ylabel('Orçamento Médio (Milhões de Dólares)')
plt.xlabel('Anos')
plt.show()
```

Portanto, da visualização anterior, percebe-se que o ano com **maior média de orçamento dos filmes de ação** foi o ano de 2007. O ano com **maior média de orçamento dos filmes de animação** foi o ano de 2008.
