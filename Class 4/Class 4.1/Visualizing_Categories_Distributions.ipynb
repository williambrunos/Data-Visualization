```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='darkgrid')
```

```python
df = sns.load_dataset('titanic')
df
```

## Categorical Scatterplot Graphic

Remember that scatterplots are used to analyse two different quantitative variables.

Categorical Scatterplot Graphics let us analyse the dispersion of a quantitative variable (y) over a categorical variable (x). We will do this with **stripplot** function.

```python
plt.figure(figsize=(10, 8))
plt.title('Ages of passengers per class')
sns.stripplot(data=df, x='class', y='age')
plt.show()
```

Notice that we can tell some conclusions about the graphic, but in a certain way the visualization is bad because of the superposition of the points. To avoid this, we can use a graphic called **swarm graphic**: it does not superposes the points.

```python
plt.figure(figsize=(10, 8))
plt.title('Ages of passengers per class')
sns.swarmplot(data=df, x='class', y='age')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Ages of passengers per sex')
sns.swarmplot(data=df, x='sex', y='age')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Countplot of people sex on the dataset')
sns.countplot(data=df, x='sex')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Ages of passengers per class')
sns.swarmplot(data=df, x='class', y='age', hue='sex')
plt.show()
```

```python
sns.catplot(data=df, x='class', y='age', hue='sex', col='survived', kind='swarm');
```

The problem of the swarm plot is the amount of space needed to no superpose the data on the graphic and the amount of colors required to plot all the categories on hue.

## Box Plots

Box plots are graphics used to visualize multiple categorical relationships, and can be veru helpful to determine in which area our dataset has the largest amount of data, determine the outliers and quartis.

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, y='age')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Age distribution')
sns.stripplot(data=df, y='age')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Age distribution')
sns.boxplot(data=df, x='class', y='age')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Age distribution per sex')
sns.boxplot(data=df, x='class', y='age', hue='sex')
plt.show()
```

Notice that categorical scatterplots let us know the relationship or scale between different categories in general. But boxplots let us analyse the 'y' variable on a better way for all the categories.

Seaborn also have the **boxenplot** function, which allow us to know the data distribution among the tails of each box graphic.

```python
plt.figure(figsize=(10, 8))
plt.title('Age distribution per sex')
sns.boxenplot(data=df, x='class', y='age', hue='sex')
plt.show()
```

```python
sns.catplot(data=df, x='class', y='age', hue='sex', col='survived', kind='boxen')
```

After all of that, we've learnt that categories scatterplots let us know the shape of the data distributtion among some class and the boxplot let us know where is the highest amount of data and the outliers.

## Violin Graphics

Violin graphics represents a probabilitie distribution mirrored curve, representing the shape of the probabilitie distribution and allow us to see where is the largest amount of data.

```python
plt.figure(figsize=(10, 8))
plt.title('Age distribution per class')
sns.violinplot(data=df, x='class', y='age')
plt.show()
```

Notice that we can see nuances in the largest intervals of data, which we cannot see on the boxplots graphics. However, violin graphics represents probabilities and not the points of data scattered.

We can also plot the violin and swarm graphics at one time, just calling the two functions.

```python
plt.figure(figsize=(10, 8))
plt.title('Ages of passengers per class')
sns.violinplot(data=df, x='class', y='age')
sns.swarmplot(data=df, x='class', y='age', color='black', size=4)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Ages of passengers per class')
sns.violinplot(data=df, x='class', y='age', hue='sex', split=True)
plt.show()
```

```python
sns.catplot(data=df, x='class', y='age', hue='sex', col='survived', kind='violin', split=True)
```

As well as in the box plot, the violin graphic doesn't let us see all the points (data) on the graphic, just de probability distribution of the data and where is the largest amount of data.
