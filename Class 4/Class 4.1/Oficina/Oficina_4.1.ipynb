## 1.

Importe as bibliotecas pandas, matplotlib e seaborn e use a palavra mágica inline, para exibir as visualizações no notebook. Depois, carregue os dados em um DataFrame. Os dados podem ser carregados pelo seaborn, utilizando a função load_dataset(‘titanic’);

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='darkgrid')
```

```python
df = sns.load_dataset('titanic')
df
```

## 2.

Crie uma visualização que mostre a distribuição do valor da passagem (“fare”), de acordo com a cidade na qual o passageiro embarcou ("embark_town"). Você pode escolher a técnica que julgar mais adequada;

```python
df['embark_town'].unique()
```

```python
plt.figure(figsize=(10, 10))
plt.title('Valor da passagem por cidade')
sns.swarmplot(data=df, x='embark_town', y='fare')
plt.show()
```

Muitos pontos estão próximos ao valor 0 no eixo y que representa o preço da passagem por conta que certas passagens, dependendo da cidade e da classe que os passageiros desejam embarcar, são muito baratas, em torno de $7,00.

Pela visualização, já podemos perceber que Queenstown possuiu grande parte das pessoas comprando passagens baratas, provavelmente desejando viajar na terceira classe. 

Cherbourg possuiu muitas vendas de tickets baratos, mas já apresentava vendas de tickets mais caros.

Southampton possuiu a maior quantidade de vendas de tickets caros, provavelmente os passageiros com maiores condições sociais compraram seus tickets desta cidade.

**Questionamento a ser feito**: as pessoas compravam os ingressos necessariamente na mesma cidade em que embarcavam? Se não, toda essa análise realizada é bem mais complexa, dado que não podemos saber exatamente a situação financeira das pessoas de cada cidade visto que elas não necessariamente compraram os tickets na mesma cidade em que embarcaram no titanic.

```python
plt.figure(figsize=(10, 10))
plt.title('Valor da passagem por cidade')
sns.boxplot(data=df, x='embark_town', y='fare')
plt.show()
```

Pela visualização do gráfico de caixas, as cidades de Queenstown e Southampton possuíram, em média, a mesma venda de tickets mais baratos, possuindo a cidade de Southampton vendas de tickets um pouco mais caros.

Os habitantes da cidade de Cherbourg compraram em média uma quantidade maior de tickets mais caros do que os habitantes das outras cidades.

```python
plt.figure(figsize=(10, 10))
plt.title('Valor da passagem por cidade')
sns.violinplot(data=df, x='embark_town', y='fare')
plt.show()
```

A visualização com gráficos de violino reforça as afirmações feitas para o gráfico de caixa.

## 2.

Crie uma visualização que mostre as distribuições do valor da passagem, de acordo com a cidade de embarque, sexo e se o passageiro sobreviveu ou não. Você pode escolher a técnica que julgar mais adequada. Que diferença você consegue notar nas distribuições dos passageiros que sobreviveram em relação aos que não sobreviveram?

```python
sns.catplot(data=df, x='embark_town', y='fare', hue='sex', col='survived', kind='violin', split=True)
```

```python
sns.catplot(data=df, x='embark_town', y='fare', hue='sex', col='survived', kind='boxen')
```

As duas visualizações nos permitem inferir que:

- Em Queenstown, as pessoas compraram tickets mais baratos, mas isso não foi algo muito relevante para a análise dado que a quantidade de pessoas que sobreviveram ou não foi praticamente a mesma.

- Em Cherbourg e em Southampton no entanto, nota-se que as pessoas que compraram tickets mais caros, provavelmente embarcando na primeira ou segunda classe, tiveram maior taxa de sobrevivência do que aquelas que compraram tickets mais baratos. Além disso, como já visto nos gráficos plotados em aula, mulheres e crianças tiveram maior taxa de sobrevivência, de acordo com os dados.
