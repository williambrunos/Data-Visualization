## 1.

Importe as bibliotecas pandas, matplotlib e seaborn e use a palavra mágica inline para exibir as visualizações no notebook. Carregue os dados em um DataFrame. Os dados podem ser carregados pelo seaborn utilizando a função load_dataset(‘mpg’).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline 
sns.set(style='darkgrid')
```

```python
df = sns.load_dataset('mpg')
df
```

## 2.

Crie uma visualização que mostre a distribuição conjunta de todas as variáveis quantitativas, diferenciando pelo país de origem (‘origin’). Quais variáveis possuem uma maior correlação?

```python
sns.pairplot(df, hue='origin')
```

Pela visualização, percebe-se que as variáveis com maiores correlações (positivas ou negativas são):
- mpg e cilindradas
- mpg e horsepower
- mpg e peso
- mpg e aceleração

Estas afirmações podem ser comprovadas analisando os gráficos com dados dispersos que são melhor adequados a uma reta de regressão na visualização par a par geral a seguir:

```python
sns.pairplot(df, kind='reg')
```

Sobre a visualização da questão 2, é possível distinguir o país de origem a partir da distribuição de uma variável? 

R: Não muito bem, pois ao analisar os gráficos de densidade na diagonal principal das duas visualizações acima, percebe-se que os carros cujo país de origem são os Estados Unidos (USA) são os carros com dados mais discrepantes dos dados cujos países de origem são da Europa ou Japão. Estes dois últimos possuem dados dos carros com valores muito parecidos, dificultando a sua diferenciação em uma visualização. A visualização que permitiria melhor distinguir o país de origem dos carros seria em relação a variável peso:

```python
plt.figure(figsize=(10, 8))
plt.title('Distribuição de pesos por país de origem do carro', fontsize=14)
sns.kdeplot(df['weight'], shade=True, hue=df['origin'])
plt.show()
```

 E de uma distribuição conjunta?

 R: Novamente, carros dos USA são mais fáceis de visualizar por possuírem dados mais discrepantes.

```python
sns.jointplot(data=df, x='displacement', y='model_year', hue='origin', height=8)
```

## 3.

Escolha uma distribuição conjunta em que as variáveis estejam fortemente correlacionadas e crie um gráfico que permita analisar a distribuição conjunta, a reta de regressão e as distribuições marginais.

```python
sns.jointplot(data=df, x='displacement', y='horsepower', kind='reg', height=8)
```

A visualização acima nos permite confirmar, o que é uma hipótese trivial, que a cilindrada do carro é fortemente correlacionada e de forma positiva com a potência do motor do mesmo carro, fato confirmado pelo 'fit' dos dados em uma reta de regressão com coeficiente angular positivo. 

No dataset, temos uma maior quantidade de carros com menos de 150 cilindradas, o que implica em uma maior quantidade de carros com potência menor ou igual a 100 cavalos, como mostram as distribuições marginais para cada variável.
