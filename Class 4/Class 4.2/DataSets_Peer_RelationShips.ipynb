```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='darkgrid')
```

# Peer Relationships Between Variables

Understand the relationship among a lot of variables in a multivariable dataset is quite a challenge, but it lead us to understand well the data and answer a lot of questions. Because of this, seaborn let us visualize those relationship among variables.

## Quick Remember

- To understand the shape of distribution of a quantitative variable, we use **histograms**.
- To understand how two quantitative variables are related, we use the **scatter** graphic. However, this kind of graphic does not let us understand the distribution of each variable. 

## Bivariate Relationship



```python
df = sns.load_dataset('iris')
df
```

```python
# No NaN values on the dataset
df.isna().sum()
```

Supose that we wanted to understand how the variables sepal_length and petal_length are distributed on the data and how these two variables are related to each other (correlation). For this, we don't need anymore to create one histogram or scatterplot for each analysis, but we can do it at one time using **jointplot** function. 

```python
sns.jointplot(data=df, x='sepal_length', y='petal_length', kind='reg', height = 8);
```

Notice that the graphic above show us the positive correlation between sepal and petal lengths and the shape of the distributions of the two variables among the data.

```python
sns.jointplot(data=df, x='sepal_length', y='petal_length', color='black', kind='hex', height=8);
```

The kind hex for the jointplot graphic divides the same in 'hex' bins of the same area, letting us to see how the data is distributed, but not if they are really related.

```python
sns.jointplot(data=df, x='sepal_length', y='petal_length', kind='kde');
plt.show()
```

The **kde** kind show us a graphic with the marginals are calculated on a distribution probability function and the scatter is showing us where it's the area with greatest probability of finding an observation (more central elipses).

## General Peer Relationship

We can also see the peer relationships between all the variables of the dataset using **pariplot** function.

```python
sns.pairplot(df)
```

The pairplot function let us see the graphics comparising two variables each time and the histogram of the variable for each element on the main diagonal. This graphic is good for general purpose, but not to investigate details on the data.

```python
sns.pairplot(df, kind='reg')
```

To see how data is distinguished by the species of iris flower, we just have to pass the name of the categoric variable 'species' for the 'hue' parameter.

```python
sns.pairplot(df, hue='species')
```
