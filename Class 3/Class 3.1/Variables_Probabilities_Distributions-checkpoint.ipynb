# Seeing Quantitative Data Distribution

It's important to understand how the data is distributed among it's variables. We can make graphs and density functions using pandas, matplotlib and seaborn.

## Histogram

Histograms are the most usual way to visualize data distributions. They are essencialy bar graph,s but the difference among them is that bar graphs calculate every observation of the variable, being mor interesting when counting discrete variables. 

Histograms plot the counts among a **range of values**, being more interesting to plot continuous values of variables.

See the matplotlib hist bar documentation here: [Hist Documentation](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.hist.html)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://leadfortaleza.com.br/ead2pcd/conteudo/tmp/myopenolat_1_103418176718354/aula/arquivos/dados.txt')
df
```

```python
def change_df_columns_names(column_name):
  column_name_as_list = column_name.split(' ')
  column_name = '_'.join(column_name_as_list)
  return column_name
```

```python
columns = df.columns
columns = list(map(change_df_columns_names, columns))
df.columns = columns 
df.head()
```

### About the Data Set

This is a brazilian house/apartment rent data set, with columns from the left to right as:
- City
- Area of the house
- Number of bathrooms
- Parking Spaces
- Floor of the house/apartment
- Allow animal
- Furnished or not
- Hoa ('condomínio')
- Rent Amount per month
- Property Tax (IPTU)
- Fire Insurence
- Total per month

## How is the Area Distributed among these houses?


```python
df['area'].describe()
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
from matplotlib import font_manager
plt.figure(figsize=(10, 5))
plt.title('Distribution of area of realty', fontsize=16)
plt.xlabel('Range of values for areas', fontsize=16)
plt.ylabel('Count of occurances', fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.hist(df['area'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Distribution of area of realty', fontsize=16)
plt.xlabel('Range of values for areas', fontsize=16)
plt.ylabel('Count of occurances', fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.hist(df['area'], bins=15)
plt.show()
```

```python
# Seeing the probability density
plt.figure(figsize=(10, 5))
plt.title('Distribution of area of realty', fontsize=16)
plt.xlabel('Range of values for areas', fontsize=16)
plt.ylabel('Probability of Occurance', fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.hist(df['area'], density = True, bins=15)
plt.show()
```

```python
# Seeing the cumulative distribution
plt.figure(figsize=(10, 5))
plt.title('Distribution of area of realty', fontsize=16)
plt.xlabel('Range of values for areas', fontsize=16)
plt.ylabel('Cumulative Probability', fontsize=16)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.hist(df['area'], density = True, cumulative = True, color='blue', bins=15)
plt.show()
```

## Density Graphs with Seaborn

Density graphs are a alternative for histograms, because it shows the data on a probabilitic way, showing the probability density curve instead of bins.

See density plot of seaborn here: [displot sns](https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot)

See KDE plot of seaborn here: [KDE plot](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot)

```python
sns.displot?
```

```python
sns.displot(data=df, x='area', kde=True);
```

```python
sns.displot(data=df, x='area', kind='kde', hue='furniture')
```

```python
sns.kdeplot?
```

```python
plt.figure(figsize=(10, 5))
plt.title('Area of realtys distribution')
plt.xlabel('Range of areas of realtys', fontsize=16)
plt.xticks(fontsize=14)
plt.ylabel('Probability of occurance', fontsize=16)
plt.yticks(fontsize=14)
sns.kdeplot(df['area'], shade=True)
plt.show()
```
