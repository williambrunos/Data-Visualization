## 1.

Importe as bibliotecas pandas, matplotlib e seaborn, usando a palavra mágica inline, para exibir as visualizações no notebook. Carregue os dados, que podem ser carregados pelo seaborn, utilizando a função load_dataset(‘mpg’), em um DataFrame.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='darkgrid')
```

```python
df = sns.load_dataset('mpg')
df
```

## Descrevendo os dados

- mpg: miles per galon: milhas por galão de combustível
- cylinders: quantidade de cilíndros
- displacement: cilindradas
- horsepower: cavalos de potência
- weight: peso
- acceleration: aceleração
- model_year: ano do modelo
- origin: país de origem
- name: nome do carro

## 2.1

Primeiro, iremos visualizar os dados de maneira geral e tentar chegar a alguma conclusão sobre a correlação entre aceleração e consumo de combustível.

```python
plt.figure(figsize=(10, 10))
plt.title('Relação entre aceleração e mpg', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('aceleração', fontsize=14)
plt.ylabel('mpg', fontsize=14)
sns.scatterplot(data=df, x='acceleration', y='mpg')
plt.show()
```

Podemos pressupor que os dados possuam uma correlação positiva, dado que a medida que a aceleração cresce, o valor de mpg também cresce. Façamos a reta de regressão para os dados de forma geral:

```python
plt.figure(figsize=(10, 10))
plt.title('Relação entre aceleração e mpg', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('aceleração', fontsize=14)
plt.ylabel('mpg', fontsize=14)
sns.regplot(data=df, x='acceleration', y='mpg')
plt.show()
```

Portanto, como visualizado no gráfico com a reta de regressão acima: há uma tendência de correlação positiva entre a aceleração e o consumo de combustível por milha, ou seja, quanto maior a aceleração do carro, maior será o consumo de combustível do mesmo.

## 2.2

Agora que sabemos que, no geral, a correlação entre a aceleração e o consumo de combustível se dá de forma positiva (quando uma cresce/decresce a uma taxa a outra cresce/decresce na mesma taxa), façamos a análise dessas correlações para os diferentes países de origens dos carros:

```python
plt.figure(figsize=(10, 10))
plt.title('Relação entre aceleração, mpg e país de origem', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('aceleração', fontsize=14)
plt.ylabel('mpg', fontsize=14)
sns.scatterplot(data=df, x='acceleration', y='mpg', hue='origin')
plt.show()
```

Portanto, é possível perceber que carros feitos nos USA estão bem distribuídos quanto a suas acelerações e consumo de combustível, mas se concentram mais em regiões de pontos com acelerações baixas/médias e, consequentemente, consumo baixo/médio de combustível.

Os carros do Japão, em média, se concentram mais em regiões com pontos de acelerações e consumos médios, sendo apenas alguns pontos com consumos altos, mesmo com acelerações consideradas médias.

Os carros da Europa, entretanto, se concentram em regiões com acelerações e consumos médios e altos, possuindo assim os carros com maiores acelerações e maiores consumos de combustível.

```python
sns.lmplot(data=df, x='acceleration', y='mpg', hue='origin', height=10);
```

Com a visualização anterior, percebe-se que as retas encontradas por meio de regressão para os dados de cada país de origem dos carros nos permitem inferir que o modelo de correlação positivo entre aceleração e consumo de combustível se mantêm de maneira positiva para os carros de diferentes origens.

```python
sns.relplot(data=df, x='acceleration', y='mpg', col='origin')
```

A visualização acima nos permite inferir que: os dados dos carros de origem nos USA e no Japão possuem maoior grau de correlação do que os dados de carros da Europa, o que pode ser também visualizado quando são analisadas as retas de regressão formadas para os três países.

## 2.3

Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year). Existe alguma relação entre as três variáveis?

```python
plt.figure(figsize=(10, 10))
plt.title('Relação entre aceleração, mpg e ano do modelo', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('aceleração', fontsize=14)
plt.ylabel('mpg', fontsize=14)
sns.scatterplot(data=df, x='acceleration', y='mpg', hue='model_year')
```

Pela visualização acima, temos uma **ideia geral** de como estão distribuídos os carros de acordo com seu ano de modelo, aceleração e consumo de combustível:

- Carros mais antigos tendem a possuir acelerações e consumos mais baixos, salvo algumas exceções em que possuem acelerações médias e consumos altos e etc.

- Carros mais novos tendem a possuir as maiores acelerações e maiores consumos de combustíveis, dado que vimos que as duas grandezas possuem correlação positiva.

Estas conclusões são reiteradas quando observamos as bolhas menores (carros mais antigos) em regiões de aceleração e consumo de combustível menores e bolhas maiores (carros mais novos) em regiões com tais grandezas mais altas.

```python
plt.figure(figsize=(10, 10))
plt.title('Relação entre aceleração, mpg e ano do modelo', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('aceleração', fontsize=14)
plt.ylabel('mpg', fontsize=14)
sns.scatterplot(data=df, x='acceleration', y='mpg', size='model_year', sizes=(15, 400))
plt.show()
```
