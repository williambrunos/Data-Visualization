```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='darkgrid')
```

```python
df = pd.read_csv('https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt')
df.head()
```

## 2.

Para o segundo e último passo, crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade, OK?

```python
plt.figure(figsize=(20, 10))
plt.title('Vendas ao longo dos meses em 3 anos', fontsize=14)
plt.yticks(fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel('Sales', fontsize=14)
plt.xlabel('Month', fontsize=12)
sns.lineplot(data=df, x='Month', y='Sales')
plt.show()
```

## Analisando as vendas por ano

```python
df_for_first_year = df[df['Month'].apply(lambda s: s[0] == '1')]
df_for_second_year = df[df['Month'].apply(lambda s: s[0] == '2')]
df_for_third_year = df[df['Month'].apply(lambda s: s[0] == '3')]
```

### Análise de vendas no primeiro ano

```python
plt.figure(figsize=(10, 8))
plt.title('Vendas por mês no primeiro ano')
sns.barplot(data=df_for_first_year, x='Month', y='Sales', edgecolor='0.4')
plt.show()
```

Percebemos que, no primeiro ano, tiveram mais vendas do shampoo nos meses de novembro e janeiro. 

### Análise das vendas no terceiro ano

```python
plt.figure(figsize=(10, 8))
plt.title('Vendas por mês no segundo ano')
sns.barplot(data=df_for_second_year, x='Month', y='Sales', edgecolor='0.4')
plt.show()
```

Desta vez, no segundo ano, os meses de maiores vendas foram outubro e dezembro.

```python
plt.figure(figsize=(10, 8))
plt.title('Vendas por mês no terceiro ano')
sns.barplot(data=df_for_third_year, x='Month', y='Sales', edgecolor='0.4')
plt.show()
```

Por fim, no terceiro ano, os meses de maiores vendas são: setembro, novembro e dezembro.

### Conclusões Gerais

Percebe-se que o fim de ano sempre é propício para as vendas do shampoo, possuindo os meses de setembro, outubro, novembro e dezembro as maiores vendas. Esse comportamento pode ser justificado pela quantidade de datas festivas (feriados como natal) ou promocionais (black friday) presentes nestes meses. 

No entanto, isto também pode ter ocorrido devido a campanhas de marketing publicitárias que acabaram alavancando as vendas do shampoo nestes como também nos demais meses, devendo o fenômeno ser investigado de forma apropriada pelos setores da empresa que produziu o shampoo.
